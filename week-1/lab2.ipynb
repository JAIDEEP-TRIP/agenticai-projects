```python
import os
import json
from dotenv import load_dotenv
from openai import OpenAI
from anthropic import Anthropic
from IPython.display import Markdown,display
from google import genai
```

```python
load_dotenv(override=True)
```

```python
openai_api = os.getenv("OPEN_API_KEY")
google_api = os.getenv("GOOGLE_API_KEY")
groq_api = os.getenv("GROQ_API_KEY")
deepseek_api = os.getenv("DEEPSEEK_API_KEY")

if openai_api:
    print("open ai key exists")
else:
    print("unable to read open ai api key")

if google_api:
    print("google key exists")
else:
    print("unable to read google api key")

if deepseek_api:
    print("deekseek key exists")
else:
    print("unable to read deepseek api key")

if groq_api:
    print("groq key exists")
else:
    print("unable to read groq api key")
```

```python
openai = OpenAI(api_key=openai_api)
```

```python
intial_task = "give me nice reasoning based question to test the reasoning and ability of a LLM"
intial_task+="answer only with question no other extra parts"
messages = [{"role":"user","content":intial_task}]
response = openai.chat.completions.create(
    model = "gpt-4o-mini",
    messages = messages

)

question = response.choices[0].message.content
print(question)
```

```python
competitors = []
answers = []
messages = [{"role": "user", "content": question}]
```

```python
model = "gpt-4o"

response = openai.chat.completions.create(
    model = model,
    messages = messages

)
competitors.append(model)
answers.append(response.choices[0].message.content)
display(Markdown(response.choices[0].message.content))
```

```python
deepseek = OpenAI(api_key=deepseek_api,base_url="https://api.deepseek.com")
model = "deepseek-v4-flash"

response = deepseek.chat.completions.create(
    model = model,
    messages = messages

)
competitors.append(model)
answers.append(response.choices[0].message.content)
display(Markdown(response.choices[0].message.content))
```

```python
client = genai.Client(api_key=google_api)
model='gemini-2.5-flash'

# Generate a basic text response
response = client.models.generate_content(
    model=model,
    contents='If a farmer has 10 apples and gives 3 to each of his 4 friends, how many apples does he have left? Consider the implications of sharing resources and how it affects both the giver and the receivers.' 
)
competitors.append(model)
answers.append(response.text)
display(Markdown(response.text))
```

```python
groq= OpenAI(api_key=groq_api,base_url="https://api.groq.com/openai/v1")
model = "openai/gpt-oss-120b"

response = groq.chat.completions.create(
    model = model,
    messages = messages

)
competitors.append(model)
answers.append(response.choices[0].message.content)
display(Markdown(response.choices[0].message.content))
```

```python
for competitor,answer in zip(competitors,answers):
    print(competitor,answer)
```

```python
together = ""
for index,answer in enumerate(answers):
    together += f"#Response from competitor : {index + 1}\n\n\n"
    together+= answer
```

```python
print(together)
```

```python
judge = f"""You are judging a competition between {len(competitors)} competitors.
Each model has been given this question:

{question}

Your job is to evaluate each response for clarity and strength of argument, and rank them in order of best to worst.
Respond with JSON, and only JSON, with the following format:
{{"results": ["best competitor number", "second best competitor number", "third best competitor number", ...]}}

Here are the responses from each competitor:

{together}

Now respond with the JSON with the ranked order of the competitors, nothing else. Do not include markdown formatting or code blocks.
just give me a list of integer """
```

```python
judge_messages = [{"role": "user", "content": judge}]
```

```python

response = openai.chat.completions.create(
    model="gpt-5-mini",
    messages=judge_messages,
)
results = response.choices[0].message.content
print(results)
```

```python
print(competitors)
```

```python
import json

parsed = json.loads(results)
actual_results = parsed["results"]
for index, result in enumerate(actual_results):
    place = int(result)
    print(f"{competitors[index]} :    {place} place")
```
