```python
question = " Pick a business are that might be worth exploring for an Agentic AI opportunity,only 1 idea just give the business area name"
messages = [{"role":"user","content":question}]
idea = openai.chat.completions.create(
    model = "gpt-4o-mini",
    messages = messages
)

industry = idea.choices[0].message.content
```

```python
question = f" find the exact pain point in the {industry} where we can implement agentic ai solutions"
messages = [{"role":"user","content":question}]
idea_2= openai.chat.completions.create(
    model = "gpt-4o-mini",
    messages = messages
)

pain_point= idea.choices[0].message.content
```

```python
question = f" propose a agentic ai solution for the above industry {industry} and the pain point is {pain_point}"
messages = [{"role":"user","content":question}]
idea_3= openai.chat.completions.create(
    model = "gpt-4o-mini",
    messages = messages
)
```
